```python
# Drop irrelevant columns

GDP = GDP.drop(["Level","Description","SeriesRefSNDQ"],axis=1)
GDP.head()
```

```python
GDP = GDP.rename(columns={"Quarter": "Quarterly-Value", "Weight":"Total-Weight", "Amount":"Total-Amount"})
GDP.head(10)
```

```python
#Getting the shape of the data (Total number of rows and columns)
#A numpy command

GDP.shape
```

```python
# Rows containing duplicate data
duplicate_rows_GDP = GDP[GDP.duplicated()]
print("Number of duplicate rows: ", duplicate_rows_GDP.describe)
```

```python
 GDP.count()
```

```python
GDP = GDP.drop_duplicates()
GDP.head(10)
```

```python
#Drop the duplicates

GDP = GDP.drop_duplicates()
GDP.count()
```

```python
#Drop the missing values
GDP = GDP.dropna()
GDP.count()
```

```python
# Find the null values 
#adding the .sum() gives you the total null values if there exist any. otherwise it gives you a boolean result.

print(GDP.isnull().sum())
```

```python
GDP.tail(30)
```

```python
#finding the outliers
#the thin vertical bars on the sides of the main value represents the deviation.

sns.boxplot(x=GDP['Total-Amount'])
```

```python
#Plotting a Histogram for the number of weight per total

GDP.Total-Amount.value_counts().nlargest(5).plot(type='bar', figsize=(20,10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.bar(GDP['Total-Weight'], GDP['Total-Amount'])
plt.xlabel('Total-Weight') 
plt.ylabel('Total-Amount')
plt.title('GDP')
plt.show()
```

```python
#find the relationship between the variables.
#N.B the closer to 1 the higher the correlation btw the pieces of data.

plt.figure(figsize=(20,10))
G = GDP.corr()
sns.heatmap(G,cmap="BrBG", annot=True)
G
```
